```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.arange(10)
```

```python
X
```

```python
plt.plot(X)
```

```python
y1 = X**2
y2 = 2*X + 3
```

```python
plt.plot(X, y1, color = 'red', linestyle = 'dashed', label = 'Parabola')
plt.plot(X, y2, color = 'green', label = 'line')
plt.xlabel('Time')
plt.ylabel('Price of Oil')
plt.title('Comodity Prices')
plt.legend(title = 'Prices:')
plt.show()
```

```python
plt.scatter(X, y1, c = 'green', label = 'plot points', marker='.')
plt.scatter(X, y2, c = 'red', label = 'Parabola', marker='^')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend(title = 'Prices:')
plt.show()
```

```python
plt.bar(['a', 'b', 'c'], [10, 20, 50])
plt.show()
```

```python
x_pos = np.array([0, 1, 2])
plt.bar(x_pos, [10, 20, 30], width = 0.4, color = 'green', label = '2007')
plt.bar(x_pos + 0.4, [15, 25, 35], width = 0.4, color = 'red', label = '2012')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend(title = 'Prices:')
plt.ylim(0, 50)
plt.xlim(-0.5, 3)
```
