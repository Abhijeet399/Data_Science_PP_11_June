```python
import keras
from keras.datasets import boston_housing
```

```python
(train_X, train_Y), (test_X, test_Y) = boston_housing.load_data()
```

```python
train_X.shape, train_Y.shape, test_X.shape, test_Y.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
MinMaxScaler?
```

```python
min_max_scaler = MinMaxScaler(feature_range = (0, 1))
```

```python
train_X = min_max_scaler.fit_transform(train_X)
test_X = min_max_scaler.fit_transform(test_X)
```

```python
min_max_scaler.fit_transform?
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
Dense?
```

```python
model = Sequential(name = 'ANN_Boston_Housing_Prediction')
model.add(Dense(200, input_dim = train_X.shape[1], activation = 'relu', name = 'Input_Layer'))
model.add(Dense(150, activation = 'relu', name = 'Layer_1'))
model.add(Dense(100, activation = 'relu', name = 'Layer_2'))
model.add(Dense(50, activation = 'relu', name = 'Layer_3'))
model.add(Dense(1, activation = 'relu', name = 'Layer_4'))
```

```python
model.compile(optimizer = 'Adam', loss = 'mse', metrics = [keras.metrics.RootMeanSquaredError()])
```

```python
model.summary()
```

```python
history = model.fit(train_X, train_Y, batch_size = 100, epochs = 1000, validation_data = (test_X, test_Y))
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
model.evaluate(test_X, test_Y)
```

```python
preds = model.predict(test_X)
```

```python
for i in range(preds.shape[0]):
    print(preds[i], test_Y[i])
```
