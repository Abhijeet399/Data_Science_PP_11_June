```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
data, label = make_blobs(n_samples = 1000, n_features = 2, centers = 5, random_state = 42)
```

```python
data.shape
```

```python
plt.scatter(data[:, 0], data[:, 1], c = label, marker='.', linewidths = 0.02)
```

```python
plt.scatter(data[:, 0], data[:, 1])
```

```python
k = 7
```

```python
color = ['green', 'blue', 'red', 'cyan', 'magenta', 'orange', 'yellow']
```

```python
clusters = {}

for i in range(k):
    center = np.random.randint(-10, 10, 2)
    points = []
    
    cluster =  {
        'center':center,
        'points':points, 
        'color':color[i]
    }
    
    clusters[i] = cluster
```

```python
clusters
```

```python
def eucledian_distance(point1, point2):
        distance = ((point1[0] - point2[0])**2 + (point1[1] - point2[1])**2)**0.5
        return distance
```

```python
def assign_points(clusters):
    
    for i in range(data.shape[0]):
        dist = []
        
        for j in range(k):
            dist.append(eucledian_distance(data[i], clusters[j]['center']))
        
        curr_cluster = np.argmin(dist)
        clusters[curr_cluster]['points'].append(data[i])
```

```python
def updation(clusters):
    
    for i in range(k):
        p = np.array(clusters[i]['points'])
        
        if p.shape[0]>0:
            mean_u = np.mean(p, axis = 0)
            clusters[i]['center'] = mean_u
            clusters[i]['points'] = []
```

```python
def plot_clusters(clusters):
    
    for i in range(k):
        p = np.array(clusters[i]['points'])
        try:
            plt.scatter(p[:, 0], p[:, 1], c = clusters[i]['color'], marker='.')
        except:
            pass
        
        center = clusters[i]['center']
        plt.scatter(center[0], center[1], marker = '^', c = 'black')
```

```python
assign_points(clusters)
plot_clusters(clusters)
```

```python
assign_points(clusters)
plot_clusters(clusters)
updation(clusters)
```
