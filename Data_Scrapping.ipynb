```python
from urllib.request import urlopen
```

```python
data = 'https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_area'
```

```python
data = urlopen(data)
```

```python
data
```

```python
print(data)
```

```python
data_html = data.read()
```

```python
print(data_html)
```

```python
from bs4 import BeautifulSoup as soup
```

```python
data_soup = soup(data_html, 'html.parser')
```

```python
print(data_soup)
```

```python
data_soup.h1
```

```python
data_soup.h2
```

```python
data_soup.find_all('h2', {})
```

```python
table = data_soup.find_all('table',{})[1]
```

```python
table = data_soup.find('table', {'class': 'wikitable sortable'})
```

```python
table
```

```python
table.find_all('tr')
```

```python
table.find_all('th')[1].text
```

```python
headings = table.find_all('th')
```

```python
col_headings = []
for i in headings:
#     print(i.text)
    try:
        col_headings.append(i.text[:-1])
    except:
        pass
```

```python
print(col_headings)
```

```python
row_data = table.find_all('tr')
```

```python
first_row = row_data[0]
```

```python
first_row
```

```python
for i in first_row:
    print(i.text)
```

```python
table_row = []

for row in row_data:
    curr_row = []
    data = row.find_all('td', {})
#     print(data)
    for i, col_data in enumerate(data):
#         print(col_data.text)
        curr_row.append(col_data.text)
    table_row.append(curr_row)
```

```python
table_row
```

```python
for row in row_data:
#     print(row.find_all('td'))
    data = row.find_all('td')
    for i, j in enumerate(data):
        print(j.text)
```

```python
csv_file = 'data_scrapped.csv'

with open(csv_file, 'w', encoding='utf_8') as file:
    headings = ','.join(col_headings)
    headings += '\n'
    file.write(headings)
    for row in table_row:
        row_str = ''
        row_str = ','.join(row)
        file.write(row_str)
```
